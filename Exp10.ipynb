```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
df=pd.read_csv('auto_mpg.csv')
sns.boxplot(df['horsepower'])
```

```python
df['horsepower']=pd.to_numeric(df['horsepower'],errors='coerce')
q1=df['horsepower'].quantile(0.25)
q3=df['horsepower'].quantile(0.75)
iqr=q3-q1 
lower_limit1=q1-(1.5*iqr)
upper_limit1=q3+(1.5*iqr)
print(upper_limit1)
print(lower_limit1)
```

```python
df[(df['horsepower']>upper_limit1)|(df['horsepower']<lower_limit1)] 
upper_cap=df['horsepower'].quantile(0.9)
lower_cap=df['horsepower'].quantile(0.1)
```

```python
df['horsepowers']=np.where(
    df['horsepower']>upper_limit1,upper_cap,np.where(df['horsepower']<=lower_limit1,lower_cap,df['horsepower']))
sns.boxplot(df['horsepowers'])
plt.show()
#after replacing outliers
```
